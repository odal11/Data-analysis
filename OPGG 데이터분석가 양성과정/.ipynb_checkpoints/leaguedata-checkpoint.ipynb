```python
import pandas as pd
import numpy as np 
```

```python
leaguedata = pd.read_csv("leaguedata.csv")
```

```python
leaguedata
```

```python
leaguedata = leaguedata[leaguedata.GP>10]
```

```python
leaguedata.sort_values(by=['GP'], axis=0, ascending = False)
```

```python
data = leaguedata
```

```python
leaguedata.columns
```

```python
for i in data.columns:
    if i in ['GP']:
        continue
    data[i]= data[i].str.replace('%','')
```

```python
data=data[['Champion','W%','Pos','GP','P%','B%','P+B%','CTR%','KDA','GD10','XPD10','CSD10','DPM','WPM', 'WCPM']]
```

```python
for i in data.columns:
    if i in ['Champion']:
        continue
    if i in ['Pos']:
        continue
    print(i)
    data[i] = pd.to_numeric(data[i])
```

```python
import pandas as pd
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split
```

```python
column = ['P%', 'B%', 'P+B%','CTR%', 'KDA', 'GD10', 'XPD10', 'CSD10','DPM', 'WPM', 'WCPM']
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[column],data['W%'], test_size = 0.2, random_state=156)
```

```python
lgbm_wrapper = LGBMRegressor(objective= "regression",
                          max_depth= 5,
                          n_estimators= 10000,
                          learning_rate= 0.001,
                          num_leaves = 31)
```

```python
LGBMRegressor
```

```python
lgbm_wrapper
```

```python
evals = [(X_test, y_test)]
lgbm_wrapper.fit(X_train, y_train, early_stopping_rounds=100, eval_metric=['rmse'], eval_set=evals, verbose=True)
```

```python
preds = lgbm_wrapper.predict(X_test)
```

```python
preds
```

```python
from lightgbm import plot_importance
import matplotlib.pyplot as plt
```

```python
plot_importance(lgbm_wrapper)
```
